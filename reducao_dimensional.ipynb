## Carregando imagem

```
from matplotlib.image import imread
from matplotlib import pyplot as plt


filename = 'razor.png'
image  = imread(filename)  
plt.imshow(image);
```

```
R, G, B = image[:,:,0], image[:,:,1], image[:,:,2]
image = 0.2989 * R + 0.5870 * G + 0.1140 * B

plt.imshow(image, cmap='gray')
plt.show()
```

```
image.shape
```

```
import numpy as np
from sklearn.decomposition import PCA

pca_all = PCA(n_components=256)
pca_all.fit(image)

plt.grid()
plt.plot(np.cumsum(pca_all.explained_variance_ratio_ * 100))
plt.xlabel('Number of components')
plt.ylabel('Explained variance')
plt.savefig('Scree plot.png')
```

Lets use 100 components

```
pca_100 = PCA(n_components=100)
image_100 = pca_100.fit_transform(image)

image_reconstructed = pca_100.inverse_transform(image_100)

plt.imshow(image_reconstructed, cmap='gray_r')
plt.axis('off')
plt.show()
```
